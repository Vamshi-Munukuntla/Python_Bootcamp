```python
def is_phone_number(p_string):
    if len(p_string) != 12:
        # not phone number size
        return False
    for i in range(0,3):
        if not p_string[i].isdecimal():
            # area code is missing
            return False 
    if p_string[3] != '-':
        # missing first dash
        return False
    for i in range(4,7):
        if not p_string[i].isdecimal():
            # missing first digits
            return False
    if p_string[7] != '-':
        # missing first dash
        return False
    for i in range(8,12):
        if not p_string[i].isdecimal():
            # missing first digits
            return False
    return True
```

```python
is_phone_number('123-124-1458')
```

```python
is_phone_number('12a-124-1458')
```

```python
is_phone_number('1234-124-1458')
```

```python
# regex that matches b/w 2 and 6 digits
def num_match(number):
    digit_regex = re.compile('(\d){2,6}?')
    mo = digit_regex.search(number)
    if mo == None:
        return ("Not Matched")
    else:
        return mo
```
