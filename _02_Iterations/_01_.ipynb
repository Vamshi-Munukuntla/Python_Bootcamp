### Problem 3

```python
student_scores = [80,60,50,65,75,55]
```

```python
def sum_score_above_average(p_student_scores):
    sum_score = 0
    number_of_students = 0
    for score in p_student_scores:
        sum_score += score
        number_of_students += 1
    average_score = sum_score/number_of_students
    sum_above_average = 0
    for score in p_student_scores:
        if score > average_score:
            sum_above_average += score
    return sum_above_average
```

```python
sum_score_above_average(student_scores)
```

## Loop over custom function

```python
def password_controller(password):
    if len(password) > 8:
        return True
    else:
        return False
    
```

```python
password_list = ['qwer','123456','5646161456','abaslfdas']
```

```python
for password in password_list:
    print(password, password_controller(password))
```

```python
def add_odd_numbers():
    total = 0
    for number in range(1,101,2):
        total += number
    return total
```

### Problem 4

```python
def add_even_numbers(start, end):
    total = 0
    for number in range(start,end+1):
        if number % 2 == 0:
            total += number
    return total
```

```python
add_even_numbers(1,101)
```
