# Project 2 - Fizz Buzz Game

```python
def fizzbuzz(start, end):
    for number in range(start, end+1):
        if number % 5  == 0 and number % 7 != 0:
            print('Fizz')
        elif number % 7  == 0 and number % 5 != 0:
            print('Buzz')
        elif number % 5  == 0 and number % 7 == 0:
            print('FizzBuzz')
        else:
            print(number)
```

```python
fizzbuzz(1,35)
```

# Project 3 - Guessing the Number

```python
import math
import random
lower = int(input('Enter Lower bound: '))
upper = int(input('Enter Upper bound: '))
number_of_chances = math.ceil(math.log(upper-lower +1, 2))
print(f"\n\tYou've only {number_of_chances} chances to guess the integer!\n")

generated_number = random.randint(lower, upper)

count = 0
while count < number_of_chances:
    count += 1
    guess = int(input('Guess a number:'))
    if generated_number == guess:
        print(f'Congratulations you did it in {count} tries.')
        break
    elif guess > generated_number:
        print("You guessed too high!")
    elif guess < generated_number:
        print("You guessed too low!")

print(f"\n The number is {generated_number}")
print('\tBetter Luck Next time!')
```

```python
import string
import random

letters = string.ascii_letters
nums = string.digits
symbols = string.punctuation

def password(a=3,b=3,c=2):
    password = ''
    if a+b+c < 8:
        print('Error: Password length should have minimum of 8 characters.')
    else:
        for i in range(a):
            password += password.join(random.choice(letters))
        for j in range(b):
            password += password.join(random.choice(nums))
        for k in range(c):
            password += password.join(random.choice(symbols))
        
        password_list  = list(password)
        random.shuffle(password_list)
        
        advanced_password = ""
        for i in password_list:
            advanced_password += i
        return f'Your password is: {advanced_password}'
```

```python
a = int(input('How many letters do you want in your password?'))
b = int(input('How many numbers do you want in your password?'))
c = int(input('How many symbols do you want in your password?'))
password(a,b,c)
```

# Project 5 - Rock, Paper, Scissors

### for single game

```python
import random
user_action = input('Please enter a choice (rock, paper, scissors):')
possible_options = ['rock','paper','scissors']
computer_action = random.choice(possible_options)
print(f'\nYou chose {user_action}, computer chose {computer_action}.')

if user_action == computer_action:
    print(f"Both players selectd {user_action}. It's a tie!")
elif user_action == 'rock':
    if computer_action == 'scissors':
        print('Rock smashes scissors! You win!')
    else:
        print('Paper covers rock! You lose!')
elif user_action == 'paper':
    if computer_action == 'scissors':
        print('Scissors cuts paper! You lose!')
    else:
        print('Paper covers rock! You win!')
elif user_action == 'scissors':
    if computer_action == 'paper':
        print('Scissors cuts paper! You win!')
    else:
        print('Rock smashes scissors! You lose!')
```

### Using Functions

```python
def select_computer_action():
    possible_options = ['rock','paper','scissors']
    computer_action = random.choice(possible_options)
    return computer_action
```

```python
def determine_winner(p_user_action, p_computer_action):
    if p_user_action == p_computer_action:
        print(f"Both players selected {p_user_action}. It's a tie!")
    elif p_user_action == 'rock':
        if p_computer_action == 'scissors':
            print('Rock smashes scissors! You win!')
        else:
            print('Paper covers rock! You lose!')
    elif p_user_action == 'paper':
        if p_computer_action == 'scissors':
            print('Scissors cuts paper! You lose!')
        else:
            print('Paper covers rock! You win!')
    elif p_user_action == 'scissors':
        if p_computer_action == 'paper':
            print('Scissors cuts paper! You win!')
        else:
            print('Rock smashes scissors! You lose!')
```

```python
while True:
    import random
    user_action = input('Please enter a choice (rock, paper, scissors):')
    computer_action = select_computer_action()
    print(f'\nYou chose {user_action}, computer chose {computer_action}.')

    determine_winner(user_action,computer_action)
            
    play_again = input('Play again (Y/N)?')
    if play_again != 'Y':
        break 
    else:
        continue
```
