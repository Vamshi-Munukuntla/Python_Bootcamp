### Multiply Dictionary Items

```python
my_dict = {"One" : 1, "Two" : 2, "Three" : 3, "Four" : 4}
```

```python
def multiply_values(p_dict):
    mul = 1
    for key in p_dict:
        mul = mul* p_dict[key]
    return mul
```

### Student Grades

```python
student_scores = {
  "John": 90,
  "Edy": 68,
  "Marry": 88, 
  "Ewan": 79,
  "Park": 62,
}
```

```python
def convert_grade(p_dict):
    student_grade = {}
    for i in student_scores:
        if student_scores[i] < 50:
            student_grade[i] = 'Fail'
        elif student_scores[i] < 64:
            student_grade[i] = 'Acceptable'
        elif student_scores[i] < 84:
            student_grade[i] = 'Good'
        else:
            student_grade[i] = 'Outstanding'
    return student_grade
    
```

```python
convert_grade(student_scores)
```

### Count Characters in a Word

```python
def count_character(word):
    output_dict = {}
    for char in word:
        if char not in output_dict:
            output_dict[char] = 1
        else:
            output_dict[char] += 1 
    return output_dict
```

```python
count_character(word)
```

```python
def items_in_store(p_dict,item):
        quantity = p_dict.get(item,-1)
        print(f'Number of {item}s in the store: {quantity}')
```

#### setdefault method

```python
def items_in_market(p_dict,item):
        quantity = p_dict.setdefault(item,0)
        print(f'Number of {item}s in the market: {quantity}')
```
