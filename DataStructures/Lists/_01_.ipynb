### Square of Items

```python
def square_list(p_list):
    for index in range(len(p_list)):
        p_list[index] = p_list[index] * p_list[index] 
    return p_list
```

```python
custom_list = [1,2,3,4,5]
square_list(custom_list)
```

#### First and Last Characters

```python
list1 = ['cbc', 'xyz', 'aba', '2332', 'abc','aa']
def count_words(list1):
    count = 0
    for index in range(len(list1)):
        if  len(list1[index]) >= 2 and list1[index][0] == list1[index][-1]:
            count += 1
    return count
```

```python
count_words(list1)
```

#### Create a list from two lists

```python
list_one = [4, 12, 16, 21, 24, 28, 32]
list_two = [5, 10, 15, 20, 25, 30, 35]
```

```python
list_three = list()
def odd_index(list_one,new_list):
    for index in range(len(list_one)):
        if index %2 == 1:
            new_list.append(list_one[index])
    return new_list
def even_index(list_two,new_list):
    for index in range(len(list_two)):
        if index %2 == 0:
            new_list.append(list_two[index])
    return new_list
```

```python
odd_index(list_one,list_three)
even_index(list_two,list_three)
```

##### Another method

```python
odd_elements = list_one[::2]
even_elements = list_two[0::2]
third_list = list()

third_list.extend(odd_elements)
third_list.extend(even_elements)
third_list
```

#### Three Equal Chunks

```python
sample_list = [21, 55, 18, 33, 24, 22, 68, 35, 79]
```

```python
length = len(sample_list)
chunk_size = int(length/3)

start = 0
end = chunk_size
for i in range(1,4):
    list_chunk = sample_list[start:end]
    print(f'Chunk_{i} = {list(reversed(list_chunk))}')
    start = end
    end += chunk_size
```
