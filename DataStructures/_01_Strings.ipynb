### Count Characters in a String

```python
def count_letter(p_sentence, p_letter):
    count = 0
    for i in p_sentence:
        if i == p_letter:
            count += 1
    return count
```

### First and Last 2 Characters

```python
def first_last_characters(word):
    if len(word) < 2:
        return ""
    else:
        word1 = word[:2] + word[-2:]
        return word1
```

### Print Pattern

```python
def print_pattern(n):
    for i in range(0,n):
        for j in range(0, i+1):
            print('*',end=' ')
        print()

    for i in range(n,0, -1):
        for j in range(0, i-1):
            print('*',end=' ')
        print()
```

```python
print_pattern(3)
```

# Project - String Formatting

```python
names = ['John', 'Edy', 'Jane', 'Kane']
scores = [90,95,80,75]
```

```python
print('{0:<10} {1:<5}'.format('Name', 'Score'))
for index in range(len(names)):
    name = names[index]
    score = scores[index]
    print('{0:<10} {1:<5}'.format(name, score))
```
