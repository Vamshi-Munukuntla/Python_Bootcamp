## Traversing Tuple - Enumerate Function

```python
my_tuple = ('one','two','three')
```

```python
for value in my_tuple:
    print(value)
```

```python
index = 0
for value in my_tuple:
    print(index,value)
    index += 1
```

```python
for index in range(len(my_tuple)):
    value = my_tuple[index]
    print(index, value)
```

```python
for index, value in enumerate(my_tuple):
    print(index, value)
```

`enumerate()` is slightly similar to dictionary.items()

```python
for index, value in enumerate(my_list):
    print(index, value)
```

```python
for value in enumerate(my_tuple):
    print(value,value[0],value[1])
```

## Search in Tuple 

```python
my_tuple = (1,2,3,4,5,6,7)
```

```python
1 in my_tuple
```

```python
10 in my_tuple
```

```python
def search_tuple(p_tuple,p_item):
    for index, value in enumerate(p_tuple):
        if value == p_item:
            return f'index of {p_item} is {index}'
    return 'The item does not exist'
```

```python
search_tuple(my_tuple,2)
```

```python
search_tuple(my_tuple,12)
```
